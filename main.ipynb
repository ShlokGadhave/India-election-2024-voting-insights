Electoral Landscape Analysis

```python
df.columns
Party_with_number_of_seats=df.groupby('Leading Party')['Leading Candidate'].count().sort_values(ascending=False).reset_index(name='count')
Party_with_number_of_seats
```

```python
plt.figure(figsize=(12,12))
sns.barplot(data=Party_with_number_of_seats,x='count',y='Leading Party',palette='viridis')
# Bharatiya Janata Party secured highest seat over all the nation 
```
