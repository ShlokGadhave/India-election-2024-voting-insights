Feature Engineering 

```python


bins = [18, 25, 40, 60, 100]

labels = ['18-25', '26-40', '41-60', '60+']

df['age_group'] = pd.cut(
    df['age'],
    bins=bins,
    labels=labels,
    include_lowest=True
)

print(df[['age', 'age_group']].head())
```

```python
#Creation of registration year column

df['registration_date'] = pd.to_datetime(df['registration_date'])
df['registration_year'] = df['registration_date'].dt.year
print(df[['registration_date', 'registration_year']].head())
```

```python
# Create a flag: missing_contact_info
df['missing_contact_info']=df["phone_number"].isnull()
print(df[['phone_number','missing_contact_info']].head())
```

```python
# Create duplicate Aadhaar flag.
df['Aadhaar Flag']=df.duplicated(subset='aadhaar_number')
print(df[['Aadhaar Flag','aadhaar_number']].head())
```

Visualizations

```python
#Voters per state 
sns.barplot(x='state',data=df,hue='gender')
plt.title('Voters per state')
plt.show()
```
